# Activation Functions
# <div dir="rtl" style="text-align: right; font-family: Tahoma, Arial, sans-serif;">توابع فعال‌سازی

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)


sigmoid = lambda x: 1 / (1 + np.exp(-x))
tanh = lambda x: np.tanh(x)
relu = lambda x: np.maximum(0, x)
leaky_relu = lambda x: np.where(x > 0, x, 0.01 * x)

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Activation Functions", fontsize=16)
axs[0, 0].plot(x, sigmoid(x), color='blue')
axs[0, 0].set_title("Sigmoid")
axs[0, 0].axhline(0, color='gray', linestyle='--')
axs[0, 0].axvline(0, color='gray', linestyle='--')
axs[0, 0].grid(True)

axs[0, 1].plot(x, tanh(x), color='green')
axs[0, 1].set_title("Tanh")
axs[0, 1].axhline(0, color='gray', linestyle='--')
axs[0, 1].axvline(0, color='gray', linestyle='--')
axs[0, 1].grid(True)

axs[1, 0].plot(x, relu(x), color='red')
axs[1, 0].set_title("ReLU")
axs[1, 0].axhline(0, color='gray', linestyle='--')
axs[1, 0].axvline(0, color='gray', linestyle='--')
axs[1, 0].grid(True)

axs[1, 1].plot(x, leaky_relu(x), color='purple')
axs[1, 1].set_title("Leaky ReLU")
axs[1, 1].axhline(0, color='gray', linestyle='--')
axs[1, 1].axvline(0, color='gray', linestyle='--')
axs[1, 1].grid(True)

# تنظیم فاصله بین پلات‌ها
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
