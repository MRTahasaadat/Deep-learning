```python
import tensorflow as tf
import matplotlib.pyplot as plt

x = tf.constant([0.5])
y_true = tf.constant([1.0])

w = tf.Variable(tf.random.normal([1]))
b = tf.Variable(tf.random.normal([1]))

learning_rate = 0.1

losses = []
predictions = []

for epoch in range(100):
    with tf.GradientTape() as tape:
        z = w * x + b
        y_pred = tf.sigmoid(z)
        loss = tf.reduce_mean(0.5 * tf.square(y_true - y_pred))

    grads = tape.gradient(loss, [w, b])
    w.assign_sub(learning_rate * grads[0])
    b.assign_sub(learning_rate * grads[1])

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: y_pred = {y_pred.numpy()[0]:.4f}, loss = {loss.numpy():.6f}")
    losses.append(loss.numpy())
    predictions.append(y_pred.numpy()[0])

```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

ax1.plot(range(100), losses, color='red')
ax1.set_title("Loss over Epochs")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.grid(True)

ax2.plot(range(100), predictions, color='blue')
ax2.set_title("Prediction (y_pred) over Epochs")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("y_pred")
ax2.grid(True)

plt.tight_layout()
plt.show()
```
